```python
def load_data(raw_data):
    scenario_all = {}
    count = 0
    for index, row in raw_data.iterrows():
        case_idx = int(row.case.split("_")[-1]) - 1
        scenario_idx = case_idx // 8
        persona_idx = case_idx % 8
        question_idx = int(row['index'])
        scenario_all[count] = llm_question_list[question_idx]+[scenarios[scenario_idx], personas[persona_idx], float(row['ratings'])]
        count += 1
    structured_data = pd.DataFrame.from_dict(scenario_all, orient='index', columns=["index", "vtype", 'higher_order_value_types', 'motivational', 'scenarios', 'personas', 'ratings'])
    return structured_data
```

### 1. Value Contents Count Distribution

```python

def _compute_percentages(group,):
    # Set type_code as index and sort
    group = group.set_index("type_code").sort_index()
    # group = group.set_index("type_value").sort_index()
    # Compute percentage of value with question group
    perc = (group["count"] / group["count"].sum()) * 100
    group["percentage"] = perc
    # Compute percentage end, centered on "Neither agree nor disagree" (type_code 0)
    # Note that we access the perc series via index which is based on 'type_code'.
    group["percentage_end"] = perc.cumsum()
    # group["percentage_end"] = perc.cumsum() - (perc[-3] + perc[-2] + perc[-1] / 2)

    # Compute percentage start by subtracting percent
    group["percentage_start"] = group["percentage_end"] - perc

    return group


def _sanity_check(new_df):
    higher_order = ['Conservation', 'Desired Values for AI Tools', 'Openness to Change',  'Self-Enhancement', 'Self-Transcendence']
    for order in higher_order:
        new_df_conversation = new_df[new_df['higher_order_value_types']==order]
        diff_set = set({-3.0, -2.0, -1.0, 0.0, 1.0, 2.0}) - set(new_df_conversation.ratings.values)
        for item in diff_set:
            new_df = new_df.append({'higher_order_value_types': order, 'ratings': item, 'count': 0}, ignore_index=True) #
    return new_df

def value_content_count_plot_data(structured_data_s1):
    
    new_df = structured_data_s1.groupby(['higher_order_value_types', 'ratings']).size().reset_index(drop=False)
    new_df.columns =['higher_order_value_types', 'ratings', 'count']
    new_df = _sanity_check(new_df)
    
    assert new_df['count'].values.sum() == len(structured_data_s1)

    new_df["order"] = new_df["higher_order_value_types"].map(
        {
            "Self-Enhancement": 1,
            "Conservation": 2,
            "Openness to Change": 3,
            "Self-Transcendence": 4,
            "Desired Values for AI Tools": 5,
        }
    )
    
    new_df["type_code"] = new_df["ratings"].map(
        {
            -3: -3,
            -2: -2,
            -1: -1,
            0: 0,
            1: 1,
            2: 2,
        }
    )
    new_df["type"] = new_df["ratings"].map(
        {
            -3: "Irrelevant",
            -2: "Strongly disagree",
            -1: "Disagree",
            0: "Neutral",
            1: "Agree",
            2: "Strongly agree",
        }
    )

    new_source = new_df.groupby("higher_order_value_types").apply(_compute_percentages).reset_index(drop=True)
    
    return new_source


import altair as alt
import pandas as pd

def plot_value_content(source, figure_title="Healthcare", legend_title="Ratings", y_axis_title="Value Types"):

    color_scale = alt.Scale(
        domain=[
            "Irrelevant",
            "Strongly disagree",
            "Disagree",
            "Neutral",
            "Agree",
            "Strongly agree",
        ],
        range=["#F8B214", "#c30d24", "#f3a583", "#cccccc", "#94c6da", "#1770ab"],
    )

    y_axis = alt.Axis(title=y_axis_title, offset=5, ticks=False, minExtent=60, domain=False)
    x_axis = alt.Axis(title="percentage_start, percentage_end")

    # title = alt.TitleParams('A centered title', anchor='middle')
    plot = alt.Chart(source).mark_bar().encode(
        x=alt.X("percentage_start:Q").axis(x_axis),
        x2="percentage_end:Q",
        y=alt.Y("higher_order_value_types:N",sort=alt.EncodingSortField(field='order')).axis(y_axis),
        color=alt.Color("type:N").title(legend_title).scale(color_scale),
    ).properties(title=figure_title)
    return plot
```

```python
import numpy as np
def compute_correlations(data_1, data_2, title_1, title_2):
    data_1_group = data_1.groupby('vtype').agg({"ratings": "mean"})
    data_2_group = data_2.groupby('vtype').agg({"ratings": "mean"})
    r = np.corrcoef(data_1_group.ratings.values, data_2_group.ratings.values)
    print(f"The Pearson product-moment correlation coefficients: {title_1} v.s. {title_2} = {r[0][1]}")
    print("group1", list(data_1_group.ratings.values))
    print("group2", list(data_2_group.ratings.values))
    return r
```

```python
import numpy as np
def compute_correlations_phi3(data_1, data_2, title_1, title_2):

    data_1_group = data_1.groupby('vtype').agg({"ratings": "mean"})
    data_2_group = data_2.groupby('vtype').agg({"ratings": "mean"})
    data_1_group_c = [float("{:.2f}".format(float(x))) for idx, x in enumerate(data_1_group.ratings.values) ] #if idx != 12
    data_2_group_c = [float("{:.2f}".format(float(x))) for idx, x in enumerate(data_2_group.ratings.values)]
    r = np.corrcoef(data_1_group_c, data_2_group_c)
    print(f"The Pearson product-moment correlation coefficients: {title_1} v.s. {title_2} = {r[0][1]}")
    print("group1", list(data_1_group_c))
    print("group2", list(data_2_group_c))
    return r
```
